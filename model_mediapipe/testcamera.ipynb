```python
import cv2
import numpy as np
import mediapipe as mp
import os
import joblib

# Utils for landmarks
mp_hands = mp.solutions.hands  # Finger and hand detection
mp_drawing = mp.solutions.drawing_utils  # Utils to draw
mp_drawing_styles = mp.solutions.drawing_styles  # Style to draw

# Initialiser le modèle MediaPipe Hands
hands = mp_hands.Hands(static_image_mode=False, min_detection_confidence=0.3)

# Dictionnaire de labels (exemple)
labels_dict = {0: 'B', 1: 'A', 2: 'C'}

# Charger le modèle (ajoutez votre code de chargement du modèle ici)
file_path = './model.joblib'
if os.path.exists(file_path):
    model = joblib.load(file_path)
    print('Model loaded successfully')
else:
    raise FileNotFoundError("model.joblib not found")

# Initialiser la capture vidéo
cap = cv2.VideoCapture(0)

while True:
    data_aux = []
    ret, frame = cap.read()
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = hands.process(frame_rgb)
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            mp_drawing.draw_landmarks(
                frame,
                hand_landmarks,
                mp_hands.HAND_CONNECTIONS,
                mp_drawing_styles.get_default_hand_landmarks_style(),
                mp_drawing_styles.get_default_hand_connections_style()
            )
            
            x_min, y_min = int(hand_landmarks.landmark[0].x * frame.shape[1]), int(hand_landmarks.landmark[0].y * frame.shape[0])
            x_max, y_max = x_min, y_min

            # Collecter les caractéristiques des points de repère
            for landmark in hand_landmarks.landmark:
                x, y = int(landmark.x * frame.shape[1]), int(landmark.y * frame.shape[0])
                x_min = min(x_min, x)
                y_min = min(y_min, y)
                x_max = max(x_max, x)
                y_max = max(y_max, y)
                data_aux.append(landmark.x)
                data_aux.append(landmark.y)

            # Vérifiez et ajustez le nombre de caractéristiques
            if len(data_aux) == model.n_features_in_:
                # Prédire le caractère
                prediction = model.predict([np.asarray(data_aux)])
                predicted_character = labels_dict[int(prediction[0])]
                
                # Dessiner un rectangle autour de la main
                cv2.rectangle(frame, (x_min - 20, y_min - 20), (x_max + 20, y_max + 20), (0, 0, 0), 4)
                
                # Afficher la prédiction sur le cadre vidéo
                cv2.putText(frame, f'{predicted_character}', (x_min - 20, y_min - 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)
            else:
                print(f"Expected {model.n_features_in_} features, but got {len(data_aux)} features.")
        
    cv2.imshow('frame', frame)
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
